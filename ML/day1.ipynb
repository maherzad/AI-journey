```python
#excersise 2 reverse list without built in reverse
temp=[]
for x in range(len(arr)-1,-1,-1):#range(start,stop,step)
  temp.append(arr[x])
temp
```

```python
#optimal way two pointers
left=0
right=len(arr)-1

while left < right:
  arr[left],arr[right]=arr[right],arr[left]
  left+=1
  right-=1
arr
```

```python
#combination of both best
seen=set()
result=[]

for num in arr:
  if num not in seen:
    seen.add(num)
    result.append(num)

print(result)
print(arr)
```

```python
#o(n)
largest=float('-inf')
secondlargest=float('-inf')
for num in arr:
  if num > largest:
    secondlargest=largest
    largest=num
  elif num > secondlargest and num != largest:
    secondlargest=num

secondlargest
```
